```python
#Descrição do conjunto de dados: "Insights do mercado de trabalho com tecnologia de IA"
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv("/content/ai_job_market_insights.csv")

# Análise exploratória
print(df.head())
print(df.describe())

# Visualização
plt.hist(df['Salary_USD'], bins=20)
plt.xlabel('Salary_USD')
plt.ylabel('Frequência')
plt.title('Distribuição de Salários')
plt.show()
```

```python
#1. Distribuição de Salários por País/Localização
# Média de salário por país
avg_salary_by_Location = df.groupby('Location')['Salary_USD'].mean().sort_values(ascending=False)
print(avg_salary_by_Location) # Changed avg_salary_by_ to avg_salary_by_Location

# Visualização: boxplot dos salários por país
plt.figure(figsize=(10, 6))
# Assuming you intend to use seaborn for the boxplot
import seaborn as sns # Importing seaborn
sns.boxplot(x='Location', y='Salary_USD', data=df) # Changed 'Country' to 'Location' to match the dataframe column name
plt.xticks(rotation=90)
plt.title('Distribuição de Salários por País')
plt.show()
```

```python
#comparar os salários médios entre as diferentes indústrias.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemplo de DataFrame (substitua isso pelo seu DataFrame real)
data = {
    'Job_Title': ['Data Scientist', 'Software Engineer', 'Data Analyst', 'DevOps Engineer'],
    'Industry': ['Tech', 'Tech', 'Finance', 'Tech'],
    'Company_Size': ['Large', 'Medium', 'Small', 'Large'],
    'Location': ['Remote', 'On-site', 'Remote', 'Remote'],
    'AI_Adoption_Level': ['High', 'Medium', 'Low', 'High'],
    'Automation_Risk': ['Low', 'Medium', 'Medium', 'Low'],
    'Required_Skills': ['Python, SQL', 'Python, AWS', 'SQL, Excel', 'Python, Docker'],
    'Salary_USD': [120000, 110000, 80000, 115000],
    'Remote_Friendly': ['Yes', 'No', 'Yes', 'Yes'],
    'Job_Growth_Projection': ['High', 'Medium', 'High', 'High']
}

df = pd.DataFrame(data)

# Calcular a média dos salários por indústria
salary_by_industry = df.groupby('Industry')['Salary_USD'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(data=salary_by_industry, x='Industry', y='Salary_USD', palette='viridis')
plt.title('Salário Médio por Indústria')
plt.xlabel('Indústria')
plt.ylabel('Salário Médio (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que você tenha seus dados em 'ai_job_market_insights.csv'
# Você também precisa carregar o DataFrame nesta célula
df = pd.read_csv("/content/ai_job_market_insights.csv") # This line is added

# Contagem de tipos de emprego
job_counts = df['Job_Title'].value_counts()

# Visualização: Gráfico de barras para as funções mais populares
plt.figure(figsize=(10, 6))
sns.barplot(y=job_counts.index[:10], x=job_counts.values[:10], palette='viridis')
plt.title('Top 10 Funções de Trabalho no Mercado de IA')
plt.xlabel('Contagem')
plt.ylabel('Função de Trabalho')
plt.show()

# Análise adicional: distribuição de setores
industry_counts = df['Industry'].value_counts()

# Visualização: Gráfico de barras para setores mais populares
plt.figure(figsize=(10, 6))
sns.barplot(y=industry_counts.index[:10], x=industry_counts.values[:10], palette='viridis')
plt.title('Top 10 Setores no Mercado de IA')
plt.xlabel('Contagem')
plt.ylabel('Setor')
plt.show()

# Análise adicional: distribuição de tamanhos de empresa
company_size_counts = df['Company_Size'].value_counts()

# Visualização: Gráfico de barras para tamanhos de empresa
plt.figure(figsize=(10, 6))
sns.barplot(y=company_size_counts.index[:10], x=company_size_counts.values[:10], palette='viridis')
plt.title('Top 10 Tamanhos de Empresa no Mercado de IA')
plt.xlabel('Contagem')
plt.ylabel('Tamanho da Empresa')
plt.show()

# Exemplo de visualização para níveis de adoção de IA
ai_adoption_counts = df['AI_Adoption_Level'].value_counts()

# Visualização: Gráfico de barras para níveis de adoção de IA
plt.figure(figsize=(10, 6))
sns.barplot(y=ai_adoption_counts.index, x=ai_adoption_counts.values, palette='viridis')
plt.title('Nível de Adoção de IA nas Empresas')
plt.xlabel('Contagem')
plt.ylabel('Nível de Adoção de IA')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criar categorias de salário
bins = [0, 50000, 100000, 150000, df['Salary_USD'].max()]
labels = ['Baixo', 'Médio', 'Alto', 'Muito Alto']
df['Salary_Category'] = pd.cut(df['Salary_USD'], bins=bins, labels=labels)

# Contagem de salários por categoria
salary_category_counts = df['Salary_Category'].value_counts()

# Visualização: gráfico de barras para categorias de salário
plt.figure(figsize=(8, 5))
sns.barplot(x=salary_category_counts.index, y=salary_category_counts.values, palette='magma')

# Títulos e rótulos
plt.title('Distribuição por Categorias de Salário')
plt.xlabel('Categoria de Salário')
plt.ylabel('Contagem')

# Exibir gráfico
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# Supondo que o DataFrame 'df' já esteja definido e carregado

# Usar One-Hot Encoding para a coluna categórica 'AI_Adoption_Level'
encoder = OneHotEncoder(drop='first')  # 'drop=first' evita a multicolinearidade
encoded_AI_level = encoder.fit_transform(df[['AI_Adoption_Level']]).toarray()

# Transformar o One-Hot Encoding em DataFrame
encoded_columns = pd.DataFrame(encoded_AI_level, columns=encoder.get_feature_names_out(['AI_Adoption_Level']))

# Usar apenas as colunas codificadas de AI_Adoption_Level
X = encoded_columns  # As variáveis explicativas são apenas as codificações de AI_Adoption_Level
y = df['Salary_USD']  # Coluna com a variável alvo

# Dividir o conjunto de dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instanciar o modelo de regressão linear
model = LinearRegression()

# Treinar o modelo com os dados de treino
model.fit(X_train, y_train)

# Fazer previsões com os dados de teste
predictions = model.predict(X_test)

# Exibir as 5 primeiras previsões
print(predictions[:5])

# Criar gráfico de linhas para comparar previsões e valores reais
plt.figure(figsize=(10, 6))

# Linha para os valores reais
plt.plot(y_test.values, label='Valores Reais', color='blue', linewidth=2)

# Linha para as previsões
plt.plot(predictions, label='Previsões', color='orange', linestyle='--', linewidth=2)

# Título e rótulos
plt.title('Comparação entre Valores Reais e Previsões', fontsize=14)
plt.xlabel('Amostras', fontsize=12)
plt.ylabel('Salário (USD)', fontsize=12)

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()

```
